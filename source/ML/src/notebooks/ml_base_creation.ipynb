```python
df1 = df[['CDM_ID','CREATED','EMERGENCY_REPORTABLE','TCA','MIN_RNG','PC']].copy()
```

```python
df2 = df[['SAT_1_ID', 'SAT_1_NAME','SAT1_OBJECT_TYPE','SAT1_RCS','SAT_1_EXCL_VOL']].copy()
```

```python
df3 = df[['SAT_2_ID', 'SAT_2_NAME','SAT2_OBJECT_TYPE','SAT2_RCS','SAT_2_EXCL_VOL']].copy()
```

```python
df2
```

```python
df2
```
